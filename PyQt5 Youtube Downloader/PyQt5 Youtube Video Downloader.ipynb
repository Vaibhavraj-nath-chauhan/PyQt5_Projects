```python
import sys
from PyQt5.QtWidgets import *
from PyQt5.QtGui import *
from PyQt5.QtCore import *
from pytube import *
from pytube import YouTube


class App(QMainWindow):

    def __init__(self):
        super().__init__()
        #adding text to application...
        self.title = 'YouTube Video Downloader'
        self.place=""
        
        #adding status baar to main window...
        self.statusBar().showMessage("                                      Welcome")
        self.statusBar().setObjectName("status_bar")
        
        
        #setting the location and size of the application...
        self.left = 700
        self.top = 250
        self.width = 400
        self.height = 550

        
        #setting the fixed size of the application...
        self.setFixedSize(self.width,self.height)
        
        
        #opening the file of css editing 
        with open("tube.css") as my_tube:
            data = my_tube.read()
        self.setStyleSheet(data)
        self.setObjectName("window")
        self.initUI()
        
    def initUI(self):
        
        #setting windows title...
        self.setWindowTitle(self.title)
        
        
        #setting application geometry...
        self.setGeometry(self.left, self.top, self.width, self.height)
        
        #adding main frame..........................................................
        self.main_window_frame = QFrame(self)
        self.main_window_frame.setObjectName("main_window")
        self.main_window_frame.move(77,80)
        
        #adding test to main_window frame...
        self.main_window_frame_text_1 = QLabel(self.main_window_frame)
        self.main_window_frame_text_1.setText("Video Downloader")
        self.main_window_frame_text_1.move(40,30)
        self.main_window_frame_text_1.setObjectName("text_1")
        
        #adding sub text...
        self.main_window_frame_text_2 = QLabel(self.main_window_frame)
        self.main_window_frame_text_2.setText("Enter link here")
        self.main_window_frame_text_2.move(30,90)
        self.main_window_frame_text_2.setObjectName("text_2")
        
        #adding input box...
        self.main_window_frame_input_1 = QLineEdit(self.main_window_frame)
        self.main_window_frame_input_1.move(30,115)
        self.main_window_frame_input_1.setObjectName("input_1")
        
        #adding sub text button...
        self.main_window_frame_text_3 = QLabel(self.main_window_frame)
        self.main_window_frame_text_3.setText("Quality")
        self.main_window_frame_text_3.move(120,150)
        self.main_window_frame_text_3.setObjectName("text_2")
        
        #adding option button...
        self.main_window_frame_option_1 = QComboBox(self.main_window_frame)
        self.main_window_frame_option_1.addItem("360p")
        self.main_window_frame_option_1.addItem("720p")
        self.main_window_frame_option_1.move(30,150)
       
        #adding button for selectiong destination...
        self.main_window_frame_button_2 = QPushButton(self.main_window_frame)
        self.main_window_frame_button_2.setText("Select")
        self.main_window_frame_button_2.clicked.connect(self.con_dir)
        self.main_window_frame_button_2.move(30,190)
        self.main_window_frame_button_2.setObjectName("button_2")
        
        #adding sub text...
        self.main_window_frame_text_4 = QLabel(self.main_window_frame)
        self.main_window_frame_text_4.setText("Destination")
        self.main_window_frame_text_4.move(110,190)
        self.main_window_frame_text_4.setObjectName("text_2")
        
        #adding entry box...
        self.main_window_frame_input_2 =QLineEdit(self.main_window_frame)
        self.main_window_frame_input_2.move(30,225)
        self.main_window_frame_input_2.setObjectName("input_1")
         
        #adding download button...
        self.main_window_frame_button_1 = QPushButton(self.main_window_frame)
        self.main_window_frame_button_1.setText("Download")
        self.main_window_frame_button_1.move(77,270)
        self.main_window_frame_button_1.setObjectName("button_1")
        self.main_window_frame_button_1.mousePressEvent = self.vi_downloader
        
    
     
#audio................................................................
        
        self.main_window_frame_2 = QFrame(self)
        self.main_window_frame_2.setObjectName("audio")
        self.main_window_frame_2.move(97,450)
        self.main_window_frame_2.mousePressEvent = self.audio_frame
        
        self.audio_text = QLabel(self.main_window_frame_2)
        self.audio_text.setText("Audio Downloader")
        self.audio_text.move(20,10)
        self.audio_text.setObjectName("audio_text")
        
        #Third frame............................................................
        self.main_window_frame_3 = QFrame(self)
        self.main_window_frame_3.setObjectName("main_window")
        self.main_window_frame_3.move(77,80)
        self.main_window_frame_3.setVisible(False)
        
        self.main_window_frame_text_10 = QLabel(self.main_window_frame_3)
        self.main_window_frame_text_10.setText("Audio Downloader")
        self.main_window_frame_text_10.move(40,30)
        self.main_window_frame_text_10.setObjectName("text_1")
        
        self.main_window_frame_text_20 = QLabel(self.main_window_frame_3)
        self.main_window_frame_text_20.setText("Enter link here")
        self.main_window_frame_text_20.move(30,90)
        self.main_window_frame_text_20.setObjectName("text_2")
        
        #adding input box...
        self.main_window_frame_input_10 = QLineEdit(self.main_window_frame_3)
        self.main_window_frame_input_10.move(30,115)
        self.main_window_frame_input_10.setObjectName("input_1")
        
        self.main_window_frame_button_20 = QPushButton(self.main_window_frame_3)
        self.main_window_frame_button_20.setText("Select")
        self.main_window_frame_button_20.clicked.connect(self.con_dir)
        self.main_window_frame_button_20.move(30,190)
        self.main_window_frame_button_20.setObjectName("button_2")
        
        #adding sub text...
        self.main_window_frame_text_40 = QLabel(self.main_window_frame_3)
        self.main_window_frame_text_40.setText("Destination")
        self.main_window_frame_text_40.move(110,190)
        self.main_window_frame_text_40.setObjectName("text_2")
        
        #adding entry box...
        self.main_window_frame_input_20 =QLineEdit(self.main_window_frame_3)
        self.main_window_frame_input_20.move(30,225)
        self.main_window_frame_input_20.setObjectName("input_1")
        
        #adding download button...
        self.main_window_frame_button_10 = QPushButton(self.main_window_frame_3)
        self.main_window_frame_button_10.setText("Download")
        self.main_window_frame_button_10.move(77,270)
        self.main_window_frame_button_10.setObjectName("button_1")
        self.main_window_frame_button_10.mousePressEvent = self.au_downloader
         
        #back button..
        self.back_1 = QLabel(self.main_window_frame_3)
        self.back_1.setTextFormat(Qt.RichText)
        self.back_1.setObjectName("back")
        self.back_1.setText("&#8592")
        self.back_1.move(10,5)
        self.back_1.mousePressEvent = self.my_back
        
        #watching the screen...
        self.show()
        
    def audio_frame(self,event):
        self.main_window_frame.setVisible(False)
        self.main_window_frame_2.setVisible(False)
        self.main_window_frame_3.setVisible(True)
        self.statusBar().showMessage("                                Audio Downloader")
        
    def my_back(self,event):
        self.main_window_frame.setVisible(True)
        self.main_window_frame_2.setVisible(True)
        self.main_window_frame_3.setVisible(False)
        self.statusBar().showMessage("                                Video Downloader")
        
        
        #defining the directory location...
    def con_dir(self):
        self.place = QFileDialog.getExistingDirectory(self,"Select Location")
        self.main_window_frame_input_2.setText(self.place)
        self.main_window_frame_input_20.setText(self.place) 
        
        
        #main function to download video...
    def downloader(self,event):
        text = str(self.main_window_frame_input_1.text())
        print(text)
        find = YouTube(text)
        print(find)
        
    def vi_downloader(self,event):
        try:
            quality = str(self.main_window_frame_option_1.currentText())
            text = str(self.main_window_frame_input_1.text())
            if text not in ["","***** Enter Url *****"]:
                video = YouTube(text).streams.filter(type="video",res=quality)[0]
                if self.place=="":
                    self.main_window_frame_input_2.setText("Select Location")
                else:
                    video.download(self.place)
                    self.main_window_frame_input_2.setText("Done")
                
            else:
                self.main_window_frame_input_1.setText("***** Enter Url *****")
        except:
            self.main_window_frame_input_1.setText("Enter  Proper URL")
        
        
    def au_downloader(self,event):
        try:
            text = str(self.main_window_frame_input_10.text())
            if text not in ["","***** Enter Url *****"]:
                audio = YouTube(text).streams.filter(type="audio")[0]
                if self.place=="":
                    self.main_window_frame_input_20.setText("Select Location")
                else:
                    audio.download(self.place)
                    self.main_window_frame_input_20.setText("Done")
        except:
            self.main_window_frame_input_10.setText("Enter Proper URL")

if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = App()
    app.exec_()
    app.quit()
```

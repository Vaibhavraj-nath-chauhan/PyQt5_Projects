```python
import PyQt5 
import sys
from PyQt5.QtCore import *
from PyQt5.QtWidgets import *
from PyQt5.QtGui import *
import os 
import pyttsx3
from time import sleep
from FaceClassifier import faceClassifier

class App(QMainWindow):

    def __init__(self):
        super().__init__()
        self.title = 'Simple Face Classifier'
        self.left = 700
        self.top = 200
        self.width = 700
        self.height = 500
        self.statusBar().showMessage("Welcome")
        self.statusBar().setObjectName("bar")
        self.setFixedSize(self.height,self.width)
        with open("face.css","r") as css:
            styleSheet = css.read()
        self.setStyleSheet(styleSheet)
        self.setObjectName("windows")
        self.engine = pyttsx3.init()
        self.initUI()
        
        
    def initUI(self):
        
        
        self.engine.say("Welcome to our program")
        self.setWindowTitle(self.title)
        self.setGeometry(self.left, self.top, self.width, self.height)
        
        
        #Phase 1
        self.login_window = QFrame(self)
        self.login_window.setObjectName("frame_1")
        self.login_window.move(55,100)
        self.login_window.setVisible(True)
        
        #Phase 2
        self.login_window_text = QLabel(self.login_window)
        self.login_window_text.setObjectName("heading")
        self.login_window_text.setText("Face Classification")
        self.login_window_text.move(120,30)
            
    
        self.text_q = QLabel(self.login_window)
        self.text_q.setObjectName("text")
        self.text_q.setText("""\tWelcome to our program\n\nWe need your face sample\n\n1. Allow Camera To Capture\n\n2. Use forefinger to classify face organs\n\n3. Press enter to close Camera""")
        self.text_q.move(30,160)
            
            
        self.login_window_button_2 = QPushButton(self.login_window)
        self.login_window_button_2.setObjectName("login_B")
        self.login_window_button_2.setText("Start")
        self.login_window_button_2.move(145,420)
        self.login_window_button_2.clicked.connect(self.register)
        self.show()
        sleep(1)
        self.engine.runAndWait()
        
    def register(self):
        cam = faceClassifier()
        self.engine.say("Request, Accepted")
        self.engine.runAndWait()
        cam.camera()
        
        
        
        
        
if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = App()
    sys.exit(app.exec_())
```
